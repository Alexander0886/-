# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
numbers = input("Введите список чисел, разделенных пробелом: ").split()
numbers = [float(num) for num in numbers]
total = 0
max_value = numbers[0]
count = 0
for num in numbers:
    total += num
    if num > max_value:
        max_value = num
    count += 1
average = total / count
print("Сумма чисел:", total)
print("Среднее значение:", average)
print("Максимальное значение:", max_value)
````

# **Задача 2.**


Напишите программу на Python, которая принимает список слов в качестве входных данных от пользователя, а затем распечатывает список в обратном порядке.

Ввод:

```
апельсин мандарин яблоко
```

Вывод:

```
яблоко мандарин апельсин
```

````python
words = input("Введите список слов, разделенных пробелом: ").split()
reversed_words = []
for i in range(len(words)-1, -1, -1):
    reversed_words.append(words[i])
print("Список слов в обратном порядке:", " ".join(reversed_words))
````

# **Задача 3**

Простой чат-бот
Создайте простого чат-бота, который отвечает на основные запросы пользователей. Программа должна использовать условные выражения для определения намерений пользователя и последующего соответствующего ответа.

Чат бот может:

- помогать складывать числа
- подсказывать фильм, музыку и тд (из заранее созданного списка)
- создавать список дел (а также выводить его, обновлять, очищать)

Интерфейс чат-бота на ваше усмотрение, это могут быть как заранее определенные команды, так и к примеру вызов команды по номеру

*Реализовать без использования функций*

Пример:

Ввод:

```
Олег, какую мызыку посоветуешь?
```

Вывод:

```
Я бы посоветовал вам: Led Zeppelin Stairway to Heaven
```

````python
films = ["Интерстеллар", "Начало", "Матрица"]
music = ["Imagine Dragons - Believer", "The Beatles - Let It Be", "Linkin Park - Numb"]
todo_list = []

print("Привет! Я простой чат-бот. Вот что я умею делать:")
print("1. Складывать числа")
print("2. Рекомендовать фильм или музыку")
print("3. Работать со списком дел (создать, показать, обновить, очистить)")
print("Для выхода введите 'выход'.")

while True:
    user_input = input("\nВведите команду или номер команды: ").lower()

    if user_input == "выход":
        print("До свидания!")
        break

    elif user_input == "1" or user_input == "сложить числа":
        try:
            num1 = float(input("Введите первое число: "))
            num2 = float(input("Введите второе число: "))
            result = num1 + num2
            print(f"Результат: {num1} + {num2} = {result}")
        except ValueError:
            print("Ошибка: введите числовое значение!")

    elif user_input == "2" or user_input == "рекомендовать":
        choice = input("Что вы хотите: фильм или музыку? ").lower()
        if choice == "фильм":
            print(f"Рекомендую посмотреть: {films[0]}")
        elif choice == "музыка":
            print(f"Рекомендую послушать: {music[0]}")
        else:
            print("Ошибка: введите 'фильм' или 'музыка'!")

    elif user_input == "3" or user_input == "список дел":
        print("Что вы хотите сделать со списком дел?")
        print("1. Показать список")
        print("2. Добавить задачу")
        print("3. Очистить список")
        action = input("Введите команду: ").lower()

        if action == "1" or action == "показать":
            if todo_list:
                print("Ваш список дел:")
                for idx, item in enumerate(todo_list, start=1):
                    print(f"{idx}. {item}")
            else:
                print("Список дел пуст.")

        elif action == "2" or action == "добавить":
            new_task = input("Введите задачу: ")
            todo_list.append(new_task)
            print(f"Задача '{new_task}' добавлена в список дел.")

        elif action == "3" or action == "очистить":
            todo_list.clear()
            print("Список дел очищен.")

        else:
            print("Ошибка: неверная команда для работы со списком дел!")

    else:
        print("Ошибка: неверная команда. Попробуйте снова.")
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
choices = ["камень", "ножницы", "бумага"]
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "Ничья"
    elif (user_choice == "камень" and computer_choice == "ножницы") or \
         (user_choice == "ножницы" and computer_choice == "бумага") or \
         (user_choice == "бумага" and computer_choice == "камень"):
        return "Пользователь"
    else:
        return "Компьютер"
user_wins = 0
computer_wins = 0

print("Игра 'Камень, ножницы, бумага' до 3 побед!")

while user_wins < 3 and computer_wins < 3:
    user_choice = input("Выберите камень, ножницы или бумага: ").lower()
    if user_choice not in choices:
        print("Некорректный выбор. Попробуйте снова.")
        continue
    computer_choice = choices [user_wins + computer_wins %3]
    print(f"Компьютер выбрал: {computer_choice}")
    result = determine_winner(user_choice, computer_choice)
    if result == "Пользователь":
        user_wins += 1
        print("Вы выиграли этот раунд!")
    elif result == "Компьютер":
        computer_wins += 1
        print("Компьютер выиграл этот раунд!")
    else:
        print("Ничья!")
    print(f"Счет: Пользователь {user_wins} - {computer_wins} Компьютер")
if user_wins == 3:
    print("Поздравляем! Вы выиграли игру!")
else:
    print("Компьютер выиграл игру! Попробуйте снова.")
````
