Описание: Создайте комплексное приложение командной строки, которое будет использоваться в качестве панели управления исследованиями космоса. Данное приложение будет обращаться к https://api.nasa.gov/ для предоставления пользователям набора информации о космосе, включая:

- Астрономическая картинка дня (APOD): Отображение APOD с пояснениями к нему.
- Фотографии с марсохода: позволяет пользователям выбирать и фильтровать фотографии с марсохода по дате и типу камеры.
- Объекты, сближающиеся с Землей (ОСЗ): Поиск и отображение информации об объектах, сближающихся с Землей, на определенную дату, включая их размеры и потенциальную опасность.
- Данные о космической погоде: Отображают последние данные о космической погоде, включая солнечные вспышки и геомагнитные бури.
Приложение должно позволять пользователям ориентироваться в этих функциях, корректно обрабатывать ошибки и обеспечивать удобство работы.

- Представление данных: Представляйте данные в четкой и организованной форме.
- Опция выхода: позволяет пользователям выходить из приложения в любое время.

```python
import requests

API_KEY = "TKma2oZWj3dQegdigUFhX1vepCVcfnwo5QwZTVeN"
BASE_URL = "https://api.nasa.gov"

def fetch_apod():

    url = f"{BASE_URL}/planetary/apod?api_key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print("\nАстрономическое изображение дня (APOD):")
        print(f"Заголовок: {data['title']}")
        print(f"Дата: {data['date']}")
        print(f"Описание: {data['explanation']}")
        print(f"URL изображения: {data['url']}")
    else:
        print("Не удалось получить изображение. Проверьте подключение или API-ключ.")

def fetch_mars_photos(date, camera):

    url = f"{BASE_URL}/mars-photos/api/v1/rovers/curiosity/photos?earth_date={date}&camera={camera}&api_key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        photos = data.get("photos", [])

        if photos:
            print(f"\nФотографии с марсохода за {date} с камеры {camera}:")
            for i, photo in enumerate(photos, start=1):
                print(f"{i}. URL: {photo['img_src']} (Земная дата: {photo['earth_date']})")
        else:
            print("Фотографии не найдены. Проверьте дату и тип камеры.")
    else:
        print("Ошибка при получении фотографий. Проверьте подключение или API-ключ.")

def fetch_neo_data(date):
    url = f"{BASE_URL}/neo/rest/v1/feed?start_date={date}&end_date={date}&api_key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        neo_objects = data["near_earth_objects"].get(date, [])

        if neo_objects:
            print(f"\nОбъекты, сближающиеся с Землей за {date}:")
            for obj in neo_objects:
                print(f"Название: {obj['name']}")
                print(f"Диаметр: от {obj['estimated_diameter']['meters']['estimated_diameter_min']:.2f} м до {obj['estimated_diameter']['meters']['estimated_diameter_max']:.2f} м")
                print(f"Опасен ли: {'Да' if obj['is_potentially_hazardous_asteroid'] else 'Нет'}")
                print(f"Скорость: {obj['close_approach_data'][0]['relative_velocity']['kilometers_per_hour']} км/ч")
                print()
        else:
            print("Нет данных о сближающихся объектах на эту дату.")
    else:
        print("Ошибка при получении данных об объектах. Проверьте подключение или API-ключ.")

def fetch_space_weather():

    url = f"{BASE_URL}/DONKI/GST?api_key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        if data:
            print("\nДанные о космической погоде:")
            for event in data:
                print(f"Начало события: {event['startTime']}")
                print(f"Описание: {event['allKpIndex']}")
                print()
        else:
            print("Нет данных о космической погоде.")
    else:
        print("Ошибка при получении данных о космической погоде.")

def main():
    while True:
        print("\nПанель управления исследованиями космоса:")
        print("1. Астрономическая картинка дня (APOD)")
        print("2. Фотографии с марсохода")
        print("3. Объекты, сближающиеся с Землей (ОСЗ)")
        print("4. Данные о космической погоде")
        print("5. Выход")

        choice = input("Выберите опцию: ").strip()

        if choice == '1':
            fetch_apod()

        elif choice == '2':
            date = input("Введите дату в формате YYYY-MM-DD: ")
            camera = input("Введите тип камеры (например, FHAZ, RHAZ, MAST, CHEMCAM, NAVCAM): ").strip().upper()
            fetch_mars_photos(date, camera)

        elif choice == '3':
            date = input("Введите дату в формате YYYY-MM-DD: ")
            fetch_neo_data(date)

        elif choice == '4':
            fetch_space_weather()

        elif choice == '5':
            print("Выход из программы.")
            break

        else:
            print("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

# Задание 2

Описание задачи

Цель этой задачи - создать скрипт на Python, который взаимодействует с API Чикагского института искусств (https://api.artic.edu/docs/) для извлечения и отображения произведений искусства. Скрипт должен позволять пользователям просматривать работы по страницам, фильтровать их по имени художника и просматривать подробную информацию о выбранных произведениях искусства. Ниже приведены требования и функциональные возможности, которые необходимо реализовать:

Требования:
Извлекать произведения искусства:

- Создайте функцию, которая извлекает список произведений искусства из API Чикагского института искусств.
Функция должна принимать параметр page для разбивки на страницы и возвращать список произведений искусства вместе с информацией о разбивке на страницы.
Фильтровать произведения искусства:

- Реализуйте функцию, которая фильтрует список произведений искусства на основе имени указанного художника. Функция должна возвращать список работ, которые соответствуют имени художника (без учета регистра).
Отображать подробную информацию об оформлении:

- Напишите функцию, которая отображает названия работ для пользователя и позволяет ему выбрать одну из них, введя соответствующий номер.
После выбора функция должна отображать подробную информацию о выбранном произведении, включая название, исполнителя, дату и носитель.
Разбивка на страницы и взаимодействие с пользователем:

- Создайте основную функцию, которая управляет выборкой произведений и взаимодействием с пользователем.

Разрешите пользователям перемещаться по страницам с произведениями искусства, выполнять фильтрацию по исполнителю или выходить из программы.

Если страниц с произведениями искусства несколько, укажите варианты перехода к следующей странице, предыдущей странице, фильтрации по исполнителю или выхода из программы.

```python
import requests

BASE_URL = "https://api.artic.edu/api/v1/artworks"

def fetch_artworks(page=1):

    url = f"{BASE_URL}?page={page}&limit=10&fields=id,title,artist_title,date_display,medium_display"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        artworks = data['data']
        pagination = data['pagination']
        return artworks, pagination
    else:
        print("Ошибка при получении данных.")
        return [], {}

def filter_artworks_by_artist(artworks, artist_name):

    return [artwork for artwork in artworks if artwork['artist_title'] and artist_name.lower() in artwork['artist_title'].lower()]

def display_artwork_details(artwork):

    print("\nДетали произведения:")
    print(f"Название: {artwork['title']}")
    print(f"Художник: {artwork['artist_title']}")
    print(f"Дата создания: {artwork['date_display']}")
    print(f"Материал: {artwork['medium_display']}")

def main():
    page = 1
    while True:
        artworks, pagination = fetch_artworks(page)

        if not artworks:
            print("Нет данных для отображения.")
            break

        print(f"\nСтраница {page}:")
        for i, artwork in enumerate(artworks, start=1):
            print(f"{i}. {artwork['title']} - {artwork['artist_title']}")

        action = input("\nВыберите действие:\n"
                       "1. Следующая страница\n"
                       "2. Предыдущая страница\n"
                       "3. Фильтр по художнику\n"
                       "4. Подробности о произведении\n"
                       "5. Выход\n"
                       "Введите номер действия: ").strip()

        if action == '1':
            if pagination.get('current_page') < pagination.get('total_pages'):
                page += 1
            else:
                print("Это последняя страница.")

        elif action == '2':
            if pagination.get('current_page') > 1:
                page -= 1
            else:
                print("Это первая страница.")

        elif action == '3':
            artist_name = input("Введите имя художника для фильтрации: ").strip()
            filtered_artworks = filter_artworks_by_artist(artworks, artist_name)

            if filtered_artworks:
                print("\nРезультаты фильтрации:")
                for i, artwork in enumerate(filtered_artworks, start=1):
                    print(f"{i}. {artwork['title']} - {artwork['artist_title']}")
            else:
                print("Не найдено произведений с указанным художником.")

        elif action == '4':
            selection = input("Введите номер произведения для просмотра деталей: ").strip()
            if selection.isdigit() and 1 <= int(selection) <= len(artworks):
                display_artwork_details(artworks[int(selection) - 1])
            else:
                print("Неверный выбор.")

        elif action == '5':
            print("Выход из программы.")
            break

        else:
            print("Неверный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
